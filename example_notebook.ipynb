```julia
using PGFPlotsX
```

```julia
x = range(-1; stop = 1, length = 20)
z = @. 1 - (abs2(x) + abs2(x'))
z[z . 0] .= -Inf
axis = @pgf Axis(
    {
        colorbar,
        "colormap/jet",
        unbounded_coords = "jump"
    },
    Plot3(
        {
            surf,
            shader = "flat",
        },
        Table(x, x, z)
    )
)
```

```julia
pgfsave("paper_example/figures/figure1.tex", axis; include_preamble=false)
```
